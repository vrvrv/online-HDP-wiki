```python
import pandas as pd
```

```python
log_path = './corpus-None-kappa-0.5-tau-1-batchsize-15/log.dat'
log = pd.read_table(log_path, sep = ' ')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(log['score'], 'bo--',linewidth = 2)
plt.plot(log['unseen.score'], 'r^-')

plt.legend(['score', 'unseen.score'], loc='lower right', fontsize = 20)
plt.xlabel("Iteration", fontsize=20)
plt.xlim(0, 50)
plt.show()
```
